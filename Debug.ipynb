```python
from ai_module.preprocessing.tf_dataset import get_val_dataset, get_dataset
from ai_module.preprocessing.features_maps_function import factory_melspec_feature

melspec_feature_mapping = factory_melspec_feature(64, 44100)
dt = get_dataset(melspec_feature_mapping, 1)
```

```python
def plot_time_series(data):
    fig = plt.figure(figsize=(14, 8))
    plt.title('Raw wave ')
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
```

```python
plot_time_series(loaded_voice)
```
